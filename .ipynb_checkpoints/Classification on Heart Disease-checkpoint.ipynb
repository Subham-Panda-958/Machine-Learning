```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
df = pd.read_csv('Datasets\heart.csv')
df.head()
```

```python
df.dtypes
df['ca'].unique()
```

```python
len(df.loc[(df['ca']=='?')
          |
          (df['thal']=='?')])
```

```python
len(df)
```

```python
X = df.iloc[:,:-1]
X.head()
```

```python
y = df['target'].copy()
y.head()
```

```python
pd.get_dummies(X, columns=['cp']).head()
```

```python
X_encoded = pd.get_dummies(X,columns=['cp','restecg','slope','thal'])
X_encoded.head()
```

```python
y.unique()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, random_state=42)
clf_dt = DecisionTreeClassifier(random_state=42)
clf_dt = clf_dt.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15,7.5))
plot_tree(clf_dt, filled = True,rounded = True, class_names = ["NO HD", "YES HD"],feature_names = X_encoded.columns);
```

```python
plot_confusion_matrix(clf_dt, X_test, y_test, display_labels = ["Does not have HD","Has HD"]);
```

```python
# predictions=clf_dt.predict(y_test)
# cm = confusion_matrix(X_test, predictions, labels=clf_dt.classes_)
# disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=clf.classes_)
# disp.plot()
```

```python
path = clf_dt.cost_complexity_pruning_path(X_train,y_train)
ccp_alphas = path.ccp_alphas
ccp_alphas = ccp_alphas[:-1]

clf_dts = []

for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    clf_dt.fit(X_train, y_train)
    clf_dts.append(clf_dt)
```

```python
train_scores = [clf_dt.score(X_train,y_train) for clf_dt in clf_dts]
test_scores = [clf_dt.score(X_test,y_test) for clf_dt in clf_dts]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs Alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker = "o", label = "train", drawstyle = "steps-post")
ax.plot(ccp_alphas, test_scores, marker = "o", label = "test", drawstyle = "steps-post")

plt.legend()
plt.show()
```

```python
clf_dt = DecisionTreeClassifier(random_state = 42, ccp_alpha = 0.027)

scores = cross_val_score(clf_dt, X_train, y_train , cv=5)
df = pd.DataFrame(data = {'tree':range(5),'accuracy':scores})

df.plot(x='tree',y='accuracy',marker = 'o',linestyle='--')
```

```python
alpha_loop_values = []

for ccp_alpha in ccp_alphas:
    clf_dt = DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    scores = cross_val_score(clf_dt, X_train, y_train, cv=5)
    alpha_loop_values.append([ccp_alpha,np.mean(scores),np.std(scores)])
    
alpha_results = pd.DataFrame(alpha_loop_values,columns=['alpha','mean_accuracy','std'])

alpha_results.plot(x='alpha',y='mean_accuracy',yerr='std',marker='o',linestyle='--')
```

```python
alpha_results[(alpha_results['alpha']>0.0126)
             &
             (alpha_results['alpha']<0.013)]
```

```python
ideal_cpp_alpha = alpha_results[(alpha_results['alpha']>0.0126)
                                &
                                (alpha_results['alpha']<0.013)]['alpha']
ideal_cpp_alpha
```

```python
ideal_cpp_alpha = float(ideal_cpp_alpha)
ideal_cpp_alpha
```

```python
clf_dt_pruned = DecisionTreeClassifier(random_state=42, ccp_alpha = ideal_cpp_alpha)
clf_dt_pruned = clf_dt_pruned.fit(X_train,y_train)
```

```python
plot_confusion_matrix(clf_dt_pruned,X_test,y_test,display_labels=["Does not have HD","Has HD"])
```

```python
plt.figure(figsize=(15,7.5))
plot_tree(clf_dt_pruned, filled = True,rounded = True, class_names = ["NO HD", "YES HD"],feature_names = X_encoded.columns);
```
